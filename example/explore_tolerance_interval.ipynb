# Tolerance Interval spread

This example shows how the spread of confidence interval of median changes with respect number of samples and confidence levels.

```python
import matplotlib.pyplot as plt
from relistats.quantile import tolerance_interval_places

percentile_levels = (75, 80, 85, 90, 95)
confidence_level = 75

n = 200

ti_bounds = [tolerance_interval_places(n, p/100, confidence_level/100) for p in percentile_levels]
ti_lo = [t[0] for t in ti_bounds]
ti_width = [t[1]-t[0]+1 for t in ti_bounds]
x_ax = range(1, n+1)
ax = plt.subplot(111)
bar_container = ax.barh(percentile_levels, ti_width, left=ti_lo, height=4, color='silver')
ax.bar_label(bar_container, labels=[f"{p}% TI @ {confidence_level}% confidence" for p in percentile_levels], label_type='center')
# for p, t in zip(percentile_levels, ti_bounds):
#      ax.barh(p, t[1]-t[0]+1, left=t[0], height=5, label=f"{p}% TI")

ax.spines[['left', 'right', 'top']].set_visible(False)
ax.get_yaxis().set_ticks([])
ax.set_xlim((1, n))
```
