# Tolerance Interval spread

This example shows how the spread of confidence interval of median changes with respect number of samples and confidence levels.

```python
import math
import matplotlib.pyplot as plt
from relistats.quantile import confidence_in_quantile

percentile_levels = (75, 80, 85, 90, 95)

n = 200

ti_lo_0 = []
ti_width_0 = []
label_0 = []
for p in percentile_levels:
    ii = math.ceil(n * (1 - p/100)/2)
    p_lo = 0.5 - p/100/2
    c_lo = confidence_in_quantile(ii, n, p_lo)
    jj = math.ceil(n - n * (1-p/100)/2)
    p_hi = 0.5 + p/100/2
    c_hi = confidence_in_quantile(jj, n, p_hi)

    print(f"{p_lo:.3f}: {ii}@{c_lo} to {p_hi:.3f}: {jj}@{c_hi}, {c_hi-c_lo}")


# ti_bounds_1 = [tolerance_interval_places(n, p/100, confidence_level/100) for p in percentile_levels]
# ti_lo_1 = [t[0] for t in ti_bounds_1]
# ti_width_1 = [t[1]-t[0]+1 for t in ti_bounds_1]
# index_1 = percentile_levels
# labels_1 = [f"{p}% TI @{confidence_level}% confidence" for p in percentile_levels]

# x_ax = range(1, n+1)
# ax = plt.subplot(111)
# bar_container = ax.barh(index_1, ti_width_1, left=ti_lo_1, height=4, color='silver')
# ax.bar_label(bar_container, labels=labels_1, label_type='center')
# # for p, t in zip(percentile_levels, ti_bounds):
# #      ax.barh(p, t[1]-t[0]+1, left=t[0], height=5, label=f"{p}% TI")

# ax.spines[['left', 'right', 'top']].set_visible(False)
# ax.get_yaxis().set_ticks([])
# ax.set_xlim((1, n))
```

```python
import matplotlib.pyplot as plt
from relistats.quantile import tolerance_interval_places

percentile_levels = (75, 80, 85, 90, 95)
confidence_level = 75

n = 200

ti_bounds_1 = [tolerance_interval_places(n, p/100, confidence_level/100) for p in percentile_levels]
ti_lo_1 = [t[0] for t in ti_bounds_1]
ti_width_1 = [t[1]-t[0]+1 for t in ti_bounds_1]
index_1 = percentile_levels
labels_1 = [f"{p}% TI @{confidence_level}% confidence" for p in percentile_levels]

x_ax = range(1, n+1)
ax = plt.subplot(111)
bar_container = ax.barh(index_1, ti_width_1, left=ti_lo_1, height=4, color='silver')
ax.bar_label(bar_container, labels=labels_1, label_type='center')
# for p, t in zip(percentile_levels, ti_bounds):
#      ax.barh(p, t[1]-t[0]+1, left=t[0], height=5, label=f"{p}% TI")

ax.spines[['left', 'right', 'top']].set_visible(False)
ax.get_yaxis().set_ticks([])
ax.set_xlim((1, n))
```

```python
import matplotlib.pyplot as plt
from relistats.quantile import assurance_interval_places

assurance_levels = (75, 80, 85, 90, 95)

n = 200

ti_bounds_3 = [assurance_interval_places(n, a/100) for a in assurance_levels]
ti_lo_3 = [t[0] for t in ti_bounds_3]
ti_width_3 = [t[1]-t[0]+1 for t in ti_bounds_3]
index_3 = assurance_levels
labels_3 = [f"{a}% Assurance Interval, {a}% TI @{a}% confidence" for a in assurance_levels]

x_ax = range(1, n+1)
ax = plt.subplot(111)
bar_container = ax.barh(assurance_levels, ti_width_3, left=ti_lo_3, height=4, color='silver')
ax.bar_label(bar_container, labels=labels_3, label_type='center')
# for p, t in zip(percentile_levels, ti_bounds):
#      ax.barh(p, t[1]-t[0]+1, left=t[0], height=5, label=f"{p}% TI")

ax.spines[['left', 'right', 'top']].set_visible(False)
ax.get_yaxis().set_ticks([])
ax.set_xlim((1, n))
```
